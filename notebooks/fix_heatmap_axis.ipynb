```python
import matplotlib.pyplot as plt
import numpy as np

# create the x coordinate
y = np.linspace(0, 50)
y = np.linspace(1, 20, 80)
# create logarithmic y coordinate
x = np.logspace(1, 4)
x = np.logspace(-3, 0, 80)

X, Y = np.meshgrid(x, y)

# create random image data to use
im_data = np.random.rand(x.size, y.size)

fig, axes = plt.subplots(
    2, 2, figsize=(8, 6), layout="constrained", sharex=True, sharey=True
)
ax = axes[0, 0]
ax.pcolormesh(X, Y, im_data)

# make y axis logarithmic
ax.set_xscale("log")
plt.show()
```
